Heart Disease: clean up

```python
# Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Name of the CSV file
heartdata = Path('Heart_Disease_Mortality_Data_Among_US_Adults__35___by_State_Territory_and_County___2019-2021.csv')
```

```python
# the correct encoding must be used to read the CSV in pandas
df = pd.read_csv(heartdata)
```

```python
# Preview of the DataFrame
df.head()
```

```python
df = df.loc[(df['Data_Value_Footnote']!="Insufficient Data") & (df['Data_Value_Type']=="Age-adjusted, Spatially Smoothed, 3-year Average Rate")]
```

```python
# Delete Columns that are not useful to our data analysis
df.drop(['DataSource', 'Data_Value_Footnote_Symbol', 'Data_Value_Footnote', 'StratificationCategory1',
        'StratificationCategory2', 'LocationID', 'Georeference', 'TopicID', 'GeographicLevel', 'Data_Value_Unit'], axis=1, inplace=True)
df.head()
```

```python
# Identify rows by count
df.count()
```

```python
unique_counts = df.nunique()
unique_counts
```

```python
rename_df = df.rename(columns={'LocationAbbr':'State',   'LocationDesc':'County', 'Data_Value':'Death_Rate(per M)',
                          'Stratification1':'Gender', 'Stratification2':'Ethnicity','Y_lat':'Latitude', 'X_lon':'Longitude'})
rename_df
```

```python
unique_counts = rename_df["Ethnicity"].value_counts()
unique_counts
```

```python
# Documenting Death rate 
df3= rename_df[rename_df["Death_Rate(per M)"].isna()== False]
df3.head()
```

```python
# Export clean data
rename_df.to_csv("cleaned_data.csv", index= False)
```
